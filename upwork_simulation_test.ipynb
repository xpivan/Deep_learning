```python
import numpy as np
import pandas as pd
from talib.abstract import *
import matplotlib.pyplot as plt
from glob import glob
import time
from IPython.core.debugger import set_trace
import datetime
```

```python
def get_data_from_csv(csv):
    data = pd.read_csv(csv)
    ma5 = SMA(data, timeperiod=5)
    ma15 = SMA(data, timeperiod=15)
    dma = ma5-ma15
    return {'data': data,
           'ma5':ma5,
           'ma15':ma15,
           'dma':dma}

def log_info_bought(stock_ticker, number_of_stock, portfolio, close, date):
    print('''stock ticker: {} 
number_of_stock: {} 
portfolio: {}
Market_value_bought: {} 
date time: {} \n'''.format(stock_ticker, number_of_stock, portfolio, close, date))

def log_info_sold(stock_ticker, number_of_stock, portfolio, close, date):
    print('''stock ticker: {}
number_of_stock: {}
portfolio: {}
Market_value_sold: {}
date time: {}\n'''.format(stock_ticker, number_of_stock, portfolio, close, date))

def log_stop_loss_short(close, date):
    print('''stop loss for short position, 
    position sold at closing price {} at {} \n'''.format(close, date))

def log_stop_loss_long(close, date):
    print('''stop loss for long position, 
    position sold at closing price {} at {} \n'''.format(close, date))
```
